Q1) Ans:
Assumptions: **1)Assume Null hyposthesis as Ho: A = B (There is no difference in diameters of cutlets between two units).**

**2)Thus Alternate hypothesis as Ha: A ≠ B (There is significant difference in diameters of cutlets between two units)**

```python
Cutlet = stats.ttest_ind(A,B)              ##### 2 Sample 2 Tail test applicable
Cutlet
```

```python
Cutlet[1]
```

 **Result**=
 Here p_value(0.4722)>aplha(Significance Value,0.05) hence **rejecting null hypothesis.**
 So their is **no significant difference** between diameters of two cutlets from different units.
