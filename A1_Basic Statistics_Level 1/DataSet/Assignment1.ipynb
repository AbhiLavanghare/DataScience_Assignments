```python
Cars=pd.read_csv("Q9_a.csv")
```

```python
Cars.head()
```

```python
Cars.drop("Index",axis=1).head()
```

**Skewness**

```python
df=pd.DataFrame(Cars)
```

```python
Car=df[["speed","dist"]]
```

```python
print(Car.skew())
```

```python
Cars["speed"].hist(figsize=(4,3))
plt.ylabel("Frequency")
plt.xlabel("Speed")
plt.title("Histogram");
```

```python
Cars["dist"].hist(figsize=(4,3))
plt.ylabel("Frequency")
plt.xlabel("Distance")
plt.title("Histogram");
```

```python
sns.distplot(Cars["dist"]);
```

```python
sns.distplot(Cars["speed"]);
```

**kurtosis**

```python
print(Car.kurtosis())
```

```python
scipy.stats.kurtosis(Cars[["speed","dist"]],axis=0,fisher=True,bias=True)
```

```python
from scipy.stats import kurtosis
print(kurtosis(Cars[["speed","dist"]], axis=0,bias=True))
```
